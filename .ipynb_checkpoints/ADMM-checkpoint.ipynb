```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from skimage.data import camera

cmap = 'gray'

im = camera().astype(np.float)[::2, ::2]
plt.imshow(im, cmap=cmap)
plt.axis('off')
plt.colorbar()
plt.title("元画像")
```

```python
from scipy.ndimage.filters import gaussian_filter

b_sig = 2
n_sig = 2

measured_im = gaussian_filter(im, b_sig) + np.random.randn( im.shape[0], im.shape[1]) * n_sig

plt.imshow(measured_im, cmap=cmap)
plt.axis('off')
plt.colorbar()
plt.title("劣化画像")
```

```python
def get_fd(im, axis):
    return np.roll(im, -1, axis=axis) - im

def get_D(im):
    return np.c_[get_fd(im, 0).flatten(), get_fd(im, 1).flatten()].reshape((im.shape[0], im.shape[1], 2))

def get_tv(im, norm):
    return np.linalg.norm(get_D(im), ord=norm, axis=2)

plt.imshow(get_tv(im, 2), cmap=cmap)
plt.axis('off')
plt.colorbar()
plt.title("TV")
```
